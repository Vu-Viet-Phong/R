## Creating a ggplot

```r
ggplot(data = mpg) +
    geom_point(mapping = aes(x = displ, y = hwy))
```

The plot shows a negative relationship between engine size (displ) and fuel efficiency (hwy). In other words, cars with big engines use more fuel.

### A scatterplot of hwy vs cyl

```r
ggplot(data = mpg) +
    geom_point(mapping = aes(x = hwy, y = cyl))
```

## Aesthetic mappings

To a two-dimensional scatterplot by mapping it to the $anesthetic$. $Anesthetic$ is a visual property of the objects in the plot. $Aesthetics$ include things like the size, shape, or color of points.

```r
ggplot(data = mpg) +
    geom_point(mapping = aes(x = displ, y = hwy, color = class))
```
